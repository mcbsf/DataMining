```python
import pandas as pd
import math
from decimal import Decimal
```

```python
#Prepara planilha de Nodes
id_list = []
label_list = []
author_list = []

id_list.append(1)
label_list.append('Subject')
author_list.append(1)
    
df = pd.read_csv('csv/topics_data', sep='\t')
print("Quantidade de tópicos: "+str(len(df)))
for i in range(len(df)):
    id_list.append('t1_'+df['id'][i])
    label_list.append('Topic')
    author_list.append(df['author'][i])
    
    
df = pd.read_csv('csv/comments_data', sep='\t')
print("Quantidade de comentários: "+str(len(df)))
for i in range(len(df)):
    id_list.append('t3_'+df['id'][i])
    label_list.append('Topic')
    author_list.append(df['author'][i])
    
d = {'id': id_list, 'Label': label_list, 'Author': author_list}
df = pd.DataFrame(data=d)
df
```

```python
df.to_csv("genphi/node.csv", encoding='utf-8')
```

```python
#Prepara planilha de Edges
source_list = []
target_list = []
type_list = []
weight_list = []
weight_log_ups_list = []
weight_ups_list = []

id_list.append(1)
label_list.append('Subject')
author_list.append(1)
    
df = pd.read_csv('csv/topics_data', sep='\t')
print("Quantidade de tópicos: "+str(len(df)))
for i in range(len(df)):
    source_list.append('t1_'+str(df['id'][i]))
    target_list.append('1')
    type_list.append('direct')
    weight_list.append(1)
    if df['ups'][i] <= 1.0:
        weight_log_ups_list.append(1)
    else:
        weight_log_ups_list.append(math.log(df['ups'][i]))
    weight_ups_list.append(df['ups'][i])
    
df = pd.read_csv('csv/comments_data', sep='\t')
print("Quantidade de comentários: "+str(len(df)))
for i in range(len(df)):
    source_list.append('t3_'+str(df['id'][i]))
    target_list.append(df['parent_id'][i])
    type_list.append('direct')
    weight_list.append(1)
    if df['ups'][i] <= 1.0:
        weight_log_ups_list.append(1)
    else:
        weight_log_ups_list.append(math.log(df['ups'][i]))
    weight_ups_list.append(df['ups'][i])
    
d = {'Source': source_list, 'Target': target_list, 'Type': type_list, 'Weight':weight_list, 'Weight_Log_Ups':weight_log_ups_list, 'Weight_Ups':weight_ups_list}
df = pd.DataFrame(data=d)
df
```

```python
df.to_csv("genphi/edge.csv", encoding='utf-8')
```
