```python
import pandas as pd
import math
from decimal import Decimal
```

```python
#Prepara planilha de Nodes
id_list = []
label_list = []
author_list = []
weight_list = []
weight_log_ups_list = []
weight_ups_list = []
sentiment_list = []
timestamp_list = []

id_list.append(1)
label_list.append('Subject')
author_list.append(1)
weight_list.append(1)
weight_log_ups_list.append(1)
weight_ups_list.append(1)
sentiment_list.append(0L)
timestamp_list.append(1) 
df = pd.read_csv('csv/topics_data.csv', sep='\t')
print("Quantidade de tópicos: "+str(len(df)))
for i in range(len(df)):
    id_list.append(df['id'][i])
    label_list.append('Topic')
    author_list.append(df['author'][i])
    weight_list.append(1)
    if df['ups'][i] <= 1.0:
        weight_log_ups_list.append(1)
    else:
        weight_log_ups_list.append(math.log(df['ups'][i]))
    weight_ups_list.append(df['ups'][i])
    sentiment_list.append(0L)
    timestamp_list.append(df['created'][i])
    
    
    
    
df = pd.read_csv('csv/comments_data.csv', sep='\t')
print("Quantidade de comentários: "+str(len(df)))
for i in range(len(df)):
    id_list.append(df['id'][i])
    label_list.append('Topic')
    author_list.append(df['author'][i])
    weight_list.append(1)
    
    sentiment_list.append(0L)
        
    if df['ups'][i] <= 1.0:
        weight_log_ups_list.append(1)
    else:
        weight_log_ups_list.append(math.log(df['ups'][i]))
    weight_ups_list.append(df['ups'][i])
    timestamp_list.append(df['created'][i])
    

d = {'id': id_list, 'Label': label_list, 'Author': author_list, 'Weight':weight_list, 'Weight_Log_Ups':weight_log_ups_list, 'Weight_Ups':weight_ups_list, 'Sentiment': sentiment_list,'timestamp': timestamp_list}
df = pd.DataFrame(data=d)
df
```

```python
df.to_csv("genphi/node.csv", encoding='utf-8', index = False)
```
